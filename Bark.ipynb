## Voice Cloning

PARAMETERS:  
audio (str) – Path to the audio file.  
speaker_id (str) – Speaker name.  
voice_dir (str) – Path to the directory to save the generate voice.

````python
!git clone https://github.com/coqui-ai/TTS.git
````

````python
from TTS.tts.configs.bark_config import BarkConfig
from TTS.tts.models.bark import Bark
import scipy
import json

# JSON 파일에서 설정 로드
config_path = '/home/giti4isgpt4/user/yeonsoo/bark_hug/generation_config.json'  # JSON 파일 경로를 정확하게 지정하세요
with open(config_path, 'r') as file:
    generation_config = json.load(file)

config = BarkConfig()
model = Bark.init_from_config(config)
model.load_checkpoint(config, checkpoint_dir="/home/giti4isgpt4/user/yeonsoo/bark_hug", eval=True)

#text = "Hello, my name is Yeonsoo. And, uh — and I like pizza. But I also have other interests such as playing tic tac toe."
text = "음성 복제 기술은 사기, 사칭, 개인 정보 보호 침해와 같은 심각한 윤리적 및 법적 문제를 야기할 수 있습니다."

# 클로닝된 speaker 사용
output_dict = model.synthesize(text, config, speaker_id="speaker_1", voice_dirs="bark_voices/")

# 로드된 설정에서 sample_rate 추출
sample_rate = generation_config['sample_rate']

# 오디오 파일 저장
audio_array = output_dict['wav']  # 오디오 데이터가 output_dict의 'audio' 키에 저장되어 있다고 가정
scipy.io.wavfile.write("bark_ko2ko.wav", rate=sample_rate, data=audio_array)
````
